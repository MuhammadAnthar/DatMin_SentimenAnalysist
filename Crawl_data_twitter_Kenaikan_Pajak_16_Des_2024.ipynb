```python
import pandas as pd
import os

# Folder tempat file CSV berada
folder_path = "tweets-data/"

# Dapatkan daftar semua file CSV dalam folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Dictionary untuk menyimpan DataFrame dengan nama df1, df2, ..., df8
dataframes = {}

# Loop untuk membaca semua file dan memberi nama DataFrame
for i, filename in enumerate(csv_files, start=1):
    file_path = os.path.join(folder_path, filename)  # Gabungkan path folder dengan nama file
    df_name = f"df{i}"  # Nama DataFrame, misalnya df1, df2, ...
    dataframes[df_name] = pd.read_csv(file_path, delimiter=",")  # Baca file CSV dan simpan ke dictionary

# Menampilkan DataFrame secara berurutan
for df_name, df in dataframes.items():
    print(f"Displaying {df_name}:")
    display(df)
```

```python
# Loop untuk mendapatkan DataFrame dari df1 hingga df8
for i in range(1, 9):  # Dari 1 hingga 8
    df_name = f"df{i}"  # Nama DataFrame secara dinamis, misalnya df1, df2, ...
    df = dataframes.get(df_name)  # Ambil DataFrame dari dictionary

    # Periksa apakah DataFrame ada
    if df is not None:
        print(f"Displaying {df_name} (Total rows: {len(df)}):")
        display(df)  # Tampilkan DataFrame
    else:
        print(f"{df_name} not found in the dictionary!")
```

```python
# Gabungkan semua DataFrame dari df1 hingga df8
df_all = pd.concat([dataframes[f"df{i}"] for i in range(1, 9)], ignore_index=True)

# Tampilkan hasil penggabungan
print(f"Total rows in merged DataFrame: {len(df_all)}")
len(df_all)
```
